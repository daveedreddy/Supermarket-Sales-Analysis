```python
#group by category
data_set[data_set["Category"]=="Dairy"]
```

```python
#total sales in dairy category
data_set[data_set["Category"]=="Dairy"]["Sales"].sum()
```

```python
#revenue generated each day
data_set["Total_revenue"]=(data_set["Quantity"]*data_set["Sales"])
```

```python
data_set
```

```python
#total revenue generated
data_set["Total_revenue"].sum()
```

```python
data_set2=data_set.groupby("Category")["Total_revenue"].sum()
```

```python
data_set2.sort_values()
```

```python
data_set2.plot(kind="bar",color="green")
plt.title("Total revenue by category")
plt.ylabel("Revenue amount")
plt.show()
```

```python
data_set["Date"]=pd.to_datetime(data_set["Date"])
```

```python
data_set3 =data_set.groupby("Date")["Total_revenue"].sum()
```

```python
data_set3
```

```python
data_set3.plot(kind="bar",color="skyblue")
plt.title("total revenue by date")
plt.ylabel("Revenue amount")
plt.show
```

```python
data_set3.plot(x="Date",y="Total_revenue",kind="line")
```
